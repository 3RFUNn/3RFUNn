# Adding Form Elements

The cell below hides quite a bit of code to calculate refactorable numbers (which are integers such that the number of divisors is itself a divisor, e.g., 9 has three divisors, 1, 3 and 9, and three divides 9). By giving the cell a #@title, you can hide the code. In addition by using the #@param tag after the initial declaration of a variable, you can expose that variable in a form for users. This helps to make notebooks more accessible and less scary for users! In this example, there are integers, string choices and a boolean check box exposed, as well as a numerical slider.

```python
#@title Refactorable Numbers

upper_bound = 100 #@param
start_digit = "Any" #@param ["Any", "1", "2", "3", "4", "5", "6", "7", "8", "9"]
must_be_odd_number = False #@param {type:"boolean"}
min_num_divisors = 30 #@param {type:"slider", min:1, max:100, step:1}

# Code for tau (number of divisors of an integer) from:
# https://rosettacode.org/wiki/Tau_function#Python

def tau(n):
    assert(isinstance(n, int) and 0 < n)
    ans, i, j = 0, 1, 1
    while i*i <= n:
        if 0 == n%i:
            ans += 1
            j = n//i
            if j != i:
                ans += 1
        i += 1
    return ans

# Note that the for loop goes to upper + 1
def show_refactorables(upper, first_digit, odd_only, min_divisors):
  for i in range(1, upper + 1):
    t = tau(i)
    if t >= min_num_divisors and (i % t) == 0:
      if not odd_only or i % 2 == 1:
        if first_digit == "Any":
          print(i)
        elif f"{i}"[0] == first_digit:
          print(i)

show_refactorables(upper_bound, start_digit, must_be_odd_number, min_num_divisors)
```

While Colab notebooks enable you to have marked up text, to show media and have simple forms, this is quite limited, and you may want to implement a more sophisticated GUI. One way to do this is to use the ipywidgets package, as we do here. Another way is to use Gradio, which has the advantage that you may be able to publish your work as a HuggingFace Space.

```python
# This cell has some helper functions which enable us to generate GUI widgets.
# For buttons, it enables us to attach a function name so that when the button
# is pressed, the function is called

import ipywidgets

def get_button(desc, on_click):
  button = ipywidgets.widgets.Button(description=desc)
  button.on_click(on_click)
  return button

def get_slider(desc, min, val, max):
  return ipywidgets.widgets.IntSlider(value=val, min=min, max=max, step=1,
    description=desc, disabled=False, continuous_update=False,
    orientation='horizontal', readout=True, readout_format='d')

def get_text_box(desc, value="", default="Blank"):
  return ipywidgets.widgets.Text(value=value, placeholder=default, description=desc, disabled=False)

def get_dropdown_list(desc, options, val=None):
  v = options[0] if val == None else val
  return ipywidgets.widgets.Dropdown(options=options, value=v, description=desc, disabled=False)

def get_label(value):
  return ipywidgets.widgets.Label(value=value)

def get_check_box(description, value):
  return ipywidgets.widgets.Checkbox(value=value, description=description, disabled=False, indent=False)
```

Colab is a great free resource, but it does have limitations that may be frustrating, especially as time gets tight for a deadline. # Python Classes

Object oriented programming is a good way to implement large code bases, and python allows this. In the following, we have a simple class with variables and methods set up. Note that you have to be very careful with the scoping of variable with classes - try and understand this example in that respect.

```python
# Here is the code for the ComputationalCreativity class. Note that all the
# function definitions should have first variable as "self", unless they are
# meant to be static.

class ComputationalCreativity:

  # Note that you can use a default value for inputs (to initiators like
  # this, byt also to any function def). Here the default is "You're the best"
  # for the motivational slogan variable.

  def __init__(self, motivational_slogan="You're the best"):
    self.motivational_slogan = motivational_slogan

  def show_me_the_slogan(self):
    print(self.motivational_slogan)

  def change_that_variable(self):
    # global variable_to_change
    variable_to_change = "HAS BEEN CHANGED NOW"

  def this_is_a_static_method(val_to_square):
    return val_to_square * val_to_square

cc = ComputationalCreativity("The best module in the whole wide world!")
cc.show_me_the_slogan()

# We can change the object's variable externally
cc.motivational_slogan = "Ain't that right."
cc.show_me_the_slogan()

# But the class cannot change a global variable, given the scope
variable_to_change = "NOT BEEN CHANGED YET"
cc.change_that_variable()

# To get it to change the "variable_to_change" variable, you need to identify
# that this is a global variable, by adding the line

# global variable_to_change

# to the top of the change_that_variable function in the class. Try it and see
# by running the cell again.

print(variable_to_change)

# Static methods can be called like this:

print(ComputationalCreativity.this_is_a_static_method(17))
```

```python
"""
Note that if you only want a class to hold data in (i.e., with no methods to
be called), you may be better off with a @dataclass, like this one below. These
can be instantiated in the same way as a normal class.
"""

from dataclasses import dataclass

@dataclass
class CCDataclass:
  motivational_slogan: "Fabuloso"
  number_of_students: int
  scores_dict: None
  is_the_best_module: bool

ccd = CCDataclass("Simply no better", 117, {"Frank": 97, "Georgina": "99"}, True)
print(ccd)
```

Here are a few things that could make python programming less annoying. In all of these, please try and work out what is happening, and ask if you don't understand. Make sure you've run the cell called "Adding Form Elements" above.

```python
# Formatting Numbers as Strings
# Note how we can limit the number of decimal places with the .3f statement

print("Formatting Numbers as Strings")
val = 0.23872634
s = f"The value of val is: {val:.3f}"
print(s + "\nAnd we can add strings together")

# Array Access
# Try and work out what's going on here with the different ways indexing arrays

print("\nArray Access")
arr = list(range(1, 1717)) # Easier to work with lists than ranges
print("arr[0] =", arr[0])
print("arr[-1] =", arr[-1])
print("arr[10:20] =", arr[10:20])
print("arr[1707:] =", arr[1707:])
print("arr[-10:] =", arr[-10:])
print("arr[-10:-5] =", arr[-10:-5])

# Substrings
# Strings can be accessed as if they are arrays of letters, which is nice

print("\nSubstrings")
# Check for existence of a substring
print("si in simon is", "si" in "simon")
# Extracting substrings
print("middle bit of simon is:", "simon"[2:4])

# Enumeration
# It's handy to have a pointer to the index as you iterate over a list.

print("\nEnumeration")
l = ["one", "two", "three", "four", "five"]
for num in l:
  print(num)
print("---")
for ind, num in enumerate(l):
  print(ind, num)

# Dictionaries
# Very handy for storage, basically a hashtable. Use keys() to get the keys

print("\nDictionaries")
dict1 = {'England': 'London', 'France': 'Paris', 'Spain': 'Madrid'}
print(f"{dict1['England']} is the capital of England")
print(list(dict1.keys()))

dict2 = {}
dict2['Australia'] = "Sydney"
print(dict2)

# List Construction
# This is a neat way to construct lists

print("\nList Construction")
squares = [i*i for i in range(0, 10)]
print("squares = ", squares)

# Sorting
# Handy for finding the best generated artefacts

print("\nSorting")
reversed_squares = sorted(squares, reverse=True) # Makes a new list
squares.sort(reverse=True) # Sorts the list in place
print("reversed_squares =", reversed_squares)
print("squares after sort =", squares)

# Sorting a list of objects

print("\nSorting a list of objects")
class ToSort:
  def __init__(self, val):
    self.val = val

clist = [ToSort(19), ToSort(7), ToSort(3), ToSort(189)]
sorted_clist = sorted(clist, key=lambda x: x.val)
print("original clist vals are:", [x.val for x in clist])
print("sorted_clist vals are:", [x.val for x in sorted_clist])

# Filters - similar to sorting, and very handy for compact code

print("\nFilters")
names = ["Simon", "Brad", "Angelina", "Brian", "Brione", "Brangelina"]
print("Names with Br in:", list(filter(lambda n: "Br" in n, names)))

ints = list(range(1, 1717))
refactorables = list(filter(lambda n: n % tau(n) == 0, ints))
print("refactorables =", refactorables)

# Compact Conditionals
# Kind of awkward, but works fine (try changing < to >)

print("\nCompact Conditionals")
check = 190 if 10 < 20 else 200
print("check =", check)

# Randomness
# The easiest way to get randomness is to use the random package

print("\nRandomness")
import random
# Note inclusivity of both end points (unlike with ranges)
options = ["optionA", "optionB", "optionC", "optionD"]
print("random.randint(0, 17) =", random.randint(0, 17))
print("random.choice(options) =", random.choice(options))
print("random.uniform(17.17, 17.1717) =", random.uniform(17.17, 17.1717))
print("bool(random.choice([True, False]) =", bool(random.choice([True, False])))
```
